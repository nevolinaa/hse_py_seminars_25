### Упорядоченные типы данных <a name="par2.1"></a>

| **Тип данных** | **Способ задания**                     | **Пример**                     | **Особенности**                                                                 |
|----------------|----------------------------------------|--------------------------------|---------------------------------------------------------------------------------|
| **Строка (str)** | `str()`<br>`''`<br>`""`                    | `"привет"`<br>`'123'`            | **Неизменяемый** тип. Состоит из символов. Поддерживает методы `.split()`, `.join()` |
| **Список (list)** | `list()`<br>`[]`                         | `[1, 2, 3]`<br>`["a", "b", "c"]` | **Изменяемый** тип. Может содержать **элементы разных типов**. Поддерживает **добавление, удаление и изменение элементов**. |
| **Кортеж (tuple)** | `tuple()`<br>`()`                        | `(1, 2, 3)`<br>`("a", "b", "c")` | **Неизменяемый** тип. В остальном по свойствам как список. |

### Операции для упорядоченных типов данных 

| **Операция**       | **Описание**                                                                 | **Пример**                                                                 |
|--------------------------|-----------------------------------------------------------------------------|----------------------------------------------------------------------------|
| **`len()`**              | Возвращает **длину** последовательности (количество элементов).                 | `len("привет")` → `6`<br>`len([1, 2, 3])` → `3`                            |
| **`in`**                 | Проверяет, **есть ли элемент в** последовательности. Возвращает `True` или `False`. | `"а" in "привет"` → `False`<br>`2 in [1, 2, 3]` → `True`                  |
| **`+`**                  | **Соединяет** две последовательности **одного типа**.                               | `"привет" + " мир"` → `"привет мир"`<br>`[1, 2] + [3, 4]` → `[1, 2, 3, 4]` |
| **Срезы `[start:end:step]`** | Позволяет извлечь **часть последовательности**.                               | `"abcdef"[1:4]` → `"bcd"`<br>`[1, 2, 3, 4][::2]` → `[1, 3]`                |

### Срезы

#### Основные параметры:

| Параметр | Описание                                                                 |
|----------|-------------------------------------------------------------------------|
| `start` | Индекс **начала** среза (включительно). Если не указан, **начинается с `0`.**   |
| `end`  | Индекс **конца** среза (не включительно). Если не указан, **идёт до конца.**    |
| `step`    | **Шаг**, с которым берутся элементы. **По умолчанию `1`.**                      |

---

#### Cтрока `"abcdef"` и список `[True, '2', 'три', 4.0, 5, 6]`:

| **Индекс** |  0  |  1  |  2  |  3  |  4  |  5  |
|--------|-----|-----|-----|-----|-----|-----|
| Элемент **строки** | 'a' | 'b' | 'c' | 'd' | 'e' | 'f' |
| Элемент **списка** | True | '2' | 'три' | 4.0 | 5 | 6 | 

---

#### Примеры срезов для строки `"abcdef"`:

| Пример                     | Результат | Объяснение                                                                 |
|----------------------------|-----------|----------------------------------------------------------------------------|
| `"abcdef"[1:4]`            | `"bcd"`   | Срез от индекса `1` (`'b'`) до индекса `4` (`'e'`, не включается).         |
| `"abcdef"[::2]`            | `"ace"`   | Срез с шагом `2` (каждый второй элемент).                                  |
| `"abcdef"[-3:-1]`          | `"de"`    | Срез от индекса `-3` (`'d'`) до индекса `-1` (`'f'`, не включается).       |
| `"abcdef"[::-1]`           | `"fedcba"`| Срез с шагом `-1` (обратный порядок).                                      |
| `"abcdef"[1:]`             | `"bcdef"` | Срез от индекса `1` до конца.                                              |
| `"abcdef"[:4]`             | `"abcd"`  | Срез от начала до индекса `4` (`'e'`, не включается).                      |
| `"abcdef"[::-2]`           | `"fdb"`   | Срез с шагом `-2` (обратный порядок, каждый второй элемент).               |

### Методы списков <a name="par2.3"></a>
 
| Метод         | Описание                                                                 | Пример                                                                 |
|---------------|-------------------------------------------------------------------------|------------------------------------------------------------------------|
| **`append()`**| Добавляет элемент в конец списка.                                       | `[1, 2].append(3)` → `[1, 2, 3]`                                      |
| **`remove()`**| Удаляет первый найденный элемент из списка.                             | `[1, 2, 3].remove(2)` → `[1, 3]`                                      |
| **`pop()`**   | Удаляет элемент по индексу и возвращает его. Если индекс не указан, удаляет последний элемент. | `[1, 2, 3].pop(1)` → `2` (список становится `[1, 3]`) |
| **`insert()`**| Вставляет элемент на указанную позицию.                                | `[1, 2].insert(1, 1.5)` → `[1, 1.5, 2]`                               |
| **`index()`** | Возвращает индекс первого найденного элемента.                         | `[1, 2, 3].index(2)` → `1`                                            |
| **`count()`** | Возвращает количество вхождений элемента в списке.                     | `[1, 2, 2, 3].count(2)` → `2`                                         |
| **`sort()`**  | Сортирует список (по умолчанию по возрастанию).                        | `[3, 1, 2].sort()` → `[1, 2, 3]`                                      |
| **`reverse()`**| Разворачивает список.                                                 | `[1, 2, 3].reverse()` → `[3, 2, 1]`                                   |

### Функция `map()` <a name="par2.4"></a>

Функция **`map()`** применяет указанную функцию к каждому элементу последовательности (например, списка) и возвращает результат в виде объекта `map`. Обычно его преобразуют в список.

#### Пример:
```
numbers = [1, 2, 3]
result = list(map(str, numbers))  # Преобразуем числа в строки
print(result)  # ["1", "2", "3"]
```

### Цикл `for` <a name="par1.2"></a>

Цикл `for` используется для **перебора элементов последовательности** (строки, списка, кортежа и т.д.).

---

#### Синтаксис:
```
for переменная in последовательность:
    # Блок кода
```

Каждый один проход по циклу называется **итерацией**  
Объект, по которому проходимся в цикле, называется **итерируемым**

### Функция `range()` 

Функция `range()` используется для **генерации последовательности чисел**. Она часто применяется в циклах `for` для выполнения определённого количества итераций.

---

#### Синтаксис вместе с for:
```
for число in range(начало, конец, шаг):
    # Блок кода
```

---

- начало с нуля
- конец не включительно

### Примеры использования `range()`:

| Описание                          | Пример                     | Вывод                     |
|----------------------------------|----------------------------|--------------------------|
| **Указан только конец.**<br>Перебор чисел от 0 до 4. | ```for i in range(5):```<br>```print(i)``` | `0`<br>`1`<br>`2`<br>`3`<br>`4` |
| **Указаны начало и конец.**<br>Перебор чисел от 2 до 5. | ```for i in range(2, 6):```<br>```print(i)``` | `2`<br>`3`<br>`4`<br>`5` |
| **Указаны начало, конец, шаг.**<br>Перебор чётных чисел. | ```for i in range(0, 10, 2):```<br>```print(i)``` | `0`<br>`2`<br>`4`<br>`6`<br>`8` |
| **Отрицательный шаг.**<br>Обратный отсчёт. | ```for i in range(5, 0, -1):```<br>```print(i)``` | `5`<br>`4`<br>`3`<br>`2`<br>`1` |
