## **Сортировки** <a name="par1"></a>

```python
# Сортировка списка чисел - по возрастанию
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]
sorted_numbers = sorted(numbers)
print(sorted_numbers)  # [1, 1, 2, 3, 4, 5, 5, 6, 9]

# Сортировка списка строк - по алфавиту
fruits = ['apple', 'banana', 'cherry', 'date']
sorted_fruits = sorted(fruits)
print(sorted_fruits)  # ['apple', 'banana', 'cherry', 'date']

# Сортировка с использованием параметра key
# Например, по длине строки
sorted_fruits_by_length = sorted(fruits, key=len)
print(sorted_fruits_by_length)  # ['date', 'apple', 'banana', 'cherry']

# Порядок сортировки: в обратном порядке
sorted_numbers_desc = sorted(numbers, reverse=True)
print(sorted_numbers_desc)  # [9, 6, 5, 5, 4, 3, 2, 1, 1]
```

## **Минимум, максимум** <a name="par2"></a>

```python
# Нахождение минимального и максимального числа
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]
min_number = min(numbers)
max_number = max(numbers)
print(min_number)  # 1
print(max_number)  # 9

# Нахождение минимальной и максимальной строки (лексикографический порядок)
fruits = ['apple', 'banana', 'cherry', 'date']
min_fruit = min(fruits)
max_fruit = max(fruits)
print(min_fruit)  # 'apple'
print(max_fruit)  # 'date'

# Использование параметра key
# Нахождение строки с минимальной и максимальной длиной
min_fruit_by_length = min(fruits, key=len)
max_fruit_by_length = max(fruits, key=len)
print(min_fruit_by_length)  # 'date'
print(max_fruit_by_length)  # 'banana'
```

## **Вложенные структуры данных** <a name="par3"></a>

### Списки списков

Каждый элемент списка сам является списком. Это удобно для представления таблиц, матриц или многомерных данных.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

#### Основные операции:
- **Доступ к элементу**: `matrix[0][1]` (это 2).
- **Итерация по элементам**:
```python
for row in matrix:
    for element in row:
        print(element)
```

### Вложенные структуры данных: словари словарей

```python
users = {
    "Alice": {"age": 25, "city": "New York"},
    "Bob": {"age": 30, "city": "London"},
    "Charlie": {"age": 22, "city": "Paris"}
}
```

Словари словарей — это структура, где значения словаря сами являются словарями. Это полезно для хранения сложных данных, например, информации о пользователях, где каждый пользователь описывается несколькими характеристиками.

#### Основные операции:
- **Доступ к элементу**: `users["Alice"]["age"]` (это 25).
- **Добавление нового пользователя**:
```python
users["Diana"] = {"age": 28, "city": "Berlin"}
```
- **Итерация по элементам**:
```python
for name, info in users.items():
    print(f"{name}: {info['age']} лет, город {info['city']}")
```

### Комбинации списков и словарей

Часто данные представляются в виде комбинаций списков и словарей. Например, список словарей или словарь списков.

#### Пример 1: Список словарей

```python
participants = [
    {"name": "Alice", "age": 25, "time": 12.5},
    {"name": "Bob", "age": 30, "time": 11.8},
    {"name": "Charlie", "age": 22, "time": 13.1}
]
```

#### Основные операции:
- **Доступ к элементу**: `participants[0]["name"]` (это "Alice").
- **Сортировка по ключу**:
```python
sorted_participants = sorted(participants, key=lambda x: x["time"])
```

#### Пример 2: Словарь списков

```python
data = {
    "names": ["Alice", "Bob", "Charlie"],
    "ages": [25, 30, 22],
    "times": [12.5, 11.8, 13.1]
}
```

#### Основные операции:
- **Доступ к элементу**: `data["names"][0]` (это "Alice").
- **Итерация по данным**:
```python
for i in range(len(data["names"])):
    print(f"{data['names'][i]}, возраст {data['ages'][i]}, время {data['times'][i]}")
```

### Вложенные структуры: словари списков словарей

Иногда данные могут быть ещё более сложными. Например, словарь, где значения — это списки словарей.


```python
teams = {
    "Team A": [
        {"name": "Alice", "score": 10},
        {"name": "Bob", "score": 15}
    ],
    "Team B": [
        {"name": "Charlie", "score": 12},
        {"name": "Diana", "score": 8}
    ]
}
```

### Когда использовать вложенные структуры?

- **Списки списков**: для таблиц, матриц, сеток
- **Словари словарей**: для хранения сложных объектов с множеством характеристик (например, пользователи, товары)
- **Список словарей**: для хранения однотипных объектов (например, участники соревнования)
- **Словарь списков**: для группировки данных (например, данные по командам)
- **Сложные вложенные структуры**: для представления иерархических данных (например, данные сотрудников компании)
