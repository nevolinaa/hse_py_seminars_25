**Зачем нужны циклы в программировании?**

- `Автоматизация повторяющихся задач` Например, для каждого студента в аудитории 1) проверить присутствие на паре 2) поставить оценку за активность
- `Экономия времени и сил` Вместо того чтобы писать один и тот же код много раз, можно использовать цикл
- `Написание более понятного и компактного кода` (логика та же, что с созданием переменных вместо использования конкретных чисел/строк и т.д.)

**Основные типы циклов в Python**

- Цикл `while` — выполняет код, пока условие истинно
- Цикл `for` — выполняет код для каждого элемента в последовательности

while True — это цикл, который выполняется бесконечно, так как условие True всегда истинно. Чтобы выйти из такого цикла, используется оператор break


### Итак, циклы  `while ` и  `for ` очень похожи
- они повторяют одни и те же действия несколько раз
- при этом мы можем использовать вместе с циклами все наши любимые цыганские фокусы :)
- и даже делать циклы в цикле (да, для  `for ` тоже!)

```
for <элемент> in <коллекция>:     # Основной цикл
    <код>            
    for <элемент> in <коллекция>: # Вложенный цикл
        <код>        
```

### Чем же отличаются цикл while и for?

| **Цикл `while`**                  | **Цикл `for`**                        |
|:---------------------------------:|:-------------------------------------:|
| Используется, когда количество итераций неизвестно | Используется для перебора коллекций |
| Требует прерывания вручную (через изменение условия или `break`)| Автоматически перебирает элементы и останавливается   |
| Пример: `while x < 10:`           | Пример: `for i in ['apple', 'banana', 'cherry']:`        |
